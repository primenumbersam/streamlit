```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D

# Euler's formula: e^(ix) = cos(x) + i*sin(x)

# Create figure and axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Generate data for the helix
t = np.linspace(0, 10*np.pi, 500)  # Parameter for the helix
x = np.cos(t)
y = np.sin(t)
z = t

# Plot the helix
line, = ax.plot(x, y, z, lw=2)

# Set axis limits
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(0, 10*np.pi)

# Set labels
ax.set_xlabel('Real')
ax.set_ylabel('Imaginary')
ax.set_zlabel('Angle (radians)')

# Set title
ax.set_title(r'Euler Formula: $e^{it} = \cos(t) + i \cdot \sin(t)$')

# Add a point to represent e^(it)
point, = ax.plot([x[0]], [y[0]], [z[0]], 'ro', markersize=8)

# Animation function
def animate(i):
    point.set_data([x[i]], [y[i]])
    point.set_3d_properties([z[i]])
    return point,

# Create animation
ani = FuncAnimation(fig, animate, frames=len(t), interval=20, blit=True)

# Add interactivity (rotate the view)
def rotate_view(event):
    ax.view_init(elev=event.ydata, azim=event.xdata)
    fig.canvas.draw()

fig.canvas.mpl_connect('motion_notify_event', rotate_view)

# Show plot
plt.show()
```

```python
# @title
# 필요한 라이브러리 설치 및 임포트
import numpy as np
import matplotlib.pyplot as plt

# x 값을 -3부터 3까지 생성
x_values = np.linspace(-3, 3, 500)

# y = x^2 값 계산
y_values = x_values**2

# 극좌표 변환: x = r cos(theta), y = r sin(theta)
r_values = np.sqrt(x_values**2 + y_values**2)
theta_values = np.arctan2(y_values, x_values)  # theta 계산

# 극좌표 그래프 그리기
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta_values, r_values, label=r'$y = x^2$ in polar coordinates')

# 그래프 포맷팅
ax.legend()
ax.set_title("Polar Plot of y = x^2 for x ∈ [-3, 3]")

# 그래프 표시
plt.show()
```

```python
# @title
# Cartesian 적분과 극좌표 적분

# 필요한 라이브러리 임포트
import numpy as np
import sympy as sp

# 심볼 정의
x = sp.Symbol('x')
theta = sp.Symbol('theta')

# 1) Cartesian 적분 계산 (x ∈ [-3,3])
integral_cartesian = sp.integrate(x**2, (x, -3, 3))

# 2) 극좌표 적분 계산 (θ ∈ [atan(-3), atan(3)])
# 극좌표 변환: R(θ) = sin(θ) / cos^2(θ)
R_expr = sp.sin(theta) / sp.cos(theta)**2

# 극좌표 적분: ∫ R^2(θ) dθ
integral_polar = sp.integrate(R_expr**2, (theta, sp.atan(-3), sp.atan(3)))

# 3) 수치적으로 계산하여 비교
numerical_cartesian = float(integral_cartesian.evalf())
numerical_polar = float(integral_polar.evalf())

# 결과 출력
numerical_cartesian, numerical_polar
```
